# Funciones

Si tengo que saludar a 30 personas, ¿voy a escribir 30 veces el mismo código?

```python
print("Hola")
print("Hola")
print("Hola")
print("Hola")
```

Las funciones sirven para agrupar instrucciones que queremos reutilizar.

## Primera función

```python
def saludar():
    print("Hola")
```

def     → define una función

saludar → nombre de la función

()      → parámetros (por ahora vacíos)

:       → inicio del bloque

Ejecutar

```python
saludar()
```

```python
saludar()
saludar()
saludar()
```

¿Y si quiero saludar a personas diferentes?

```python
def saludar(nombre):
    print("Hola", nombre)
```

```python
saludar("Juan")
saludar("Ana")
saludar("Pedro")
```

# Sumar

```python
def sumar(a,b):
    print(a+b)
```

```python
sumar(3,4)
```

¿Qué diferencia hay entre mostrar un resultado y devolverlo?

`sumar(3,4)`  

Muestra:

`7`

pero no guarda nada.

## Return

```python
def sumar(a,b):
    return a+b
```

```python
resultado = sumar(3,4)
```

```python
print(resultado)
```

**Analogía de la cafetería**

Parámetros → ingredientes

Función → cocinero

Return → plato terminado

*Si el cocinero solo grita:*

*"La comida está lista"*

*eso sería print*


`Si te entrega el plato:`

`eso sería return.`

```python
def area_rectangulo(base, altura):
    return base * altura

area = area_rectangulo(5,3)

print(area)
```
