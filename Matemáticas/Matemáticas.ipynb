# Resolver ecuaciones lineales

Importar la librería `sympy`

```python
import sympy as sp
```

```python
# Declaramos cual es la variable
x = sp.symbols('x')
```

Se revolverá la ecuación

**7x - 3 = 4x + 9**

```python
# Ejemplo: 7x - 3 = 4x + 9

# Declaramos la ecuación
ecuacion = sp.Eq(7*x - 3, 4*x + 9)

# Resolvemos
solucion = sp.solve(ecuacion, x)
```

```python
# Imprimir el resultado
print(f"La solución es: {solucion[0]}")
```
