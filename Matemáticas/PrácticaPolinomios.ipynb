# Práctica de raíces de un polinomio

## Códigos

### Código para polinomios de segundo grado


```python
# Bloque 1: Importaciones y definición de funciones
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Función del polinomio
def f(x, a, b, c):
    """
    Calcula el valor del polinomio de segundo grado
    f(x) = ax² + bx + c
    """
    return a * x**2 + b * x + c

def calcular_raices(a, b, c):
    """
    Calcula las raíces del polinomio de segundo grado
    Retorna: (raiz1, raiz2) o (None, None) si no hay raíces reales
    """
    if a == 0:
        # Si a=0, es una ecuación lineal
        if b != 0:
            return [-c/b], None
        return None, None

    discriminante = b**2 - 4*a*c

    if discriminante < 0:
        # No hay raíces reales
        return None, None
    elif discriminante == 0:
        # Una raíz doble
        raiz = -b / (2*a)
        return raiz, raiz
    else:
        # Dos raíces reales diferentes
        raiz1 = (-b + np.sqrt(discriminante)) / (2*a)
        raiz2 = (-b - np.sqrt(discriminante)) / (2*a)
        return raiz1, raiz2

def graficar_polinomio(a=1.0, b=0.0, c=0.0):
    """
    Función que grafica el polinomio con los coeficientes dados
    y marca las raíces en color morado
    """
    # Crear datos para graficar
    x = np.linspace(-10, 10, 1000)
    y = f(x, a, b, c)

    # Crear la gráfica
    plt.figure(figsize=(10, 6))

    # Graficar el polinomio
    plt.plot(x, y, linewidth=2, color='blue', label=f'f(x) = {a:.2f}x² + {b:.2f}x + {c:.2f}')

    # Configurar el gráfico
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Polinomio de segundo grado: ax² + bx + c')
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)

    # Calcular y mostrar el vértice
    if a != 0:
        x_v = -b / (2*a)
        y_v = f(x_v, a, b, c)
        plt.plot(x_v, y_v, 'ro', markersize=8, label=f'Vértice: ({x_v:.2f}, {y_v:.2f})')

    # Calcular y mostrar las raíces (puntos morados)
    raiz1, raiz2 = calcular_raices(a, b, c)

    # Lista para almacenar las raíces
    raices = []
    if raiz1 is not None:
        raices.append(raiz1)
    if raiz2 is not None and raiz2 != raiz1:
        raices.append(raiz2)

    # Graficar las raíces
    if raices:
        for raiz in raices:
            # Verificar que la raíz esté dentro del rango de visualización
            if -10 <= raiz <= 10:
                y_raiz = f(raiz, a, b, c)  # Esto debería ser ~0
                plt.plot(raiz, y_raiz, 'm^', markersize=10,
                        label='Raíces' if raiz == raices[0] else "")

                # Mostrar el valor de la raíz en el gráfico
                plt.annotate(f'x = {raiz:.2f}',
                           xy=(raiz, y_raiz),
                           xytext=(10, 10),
                           textcoords='offset points',
                           color='purple',
                           fontsize=10,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        # Agregar información sobre las raíces
        if len(raices) == 1:
            info_raices = f'Raíz: {raices[0]:.2f} (doble)'
        else:
            info_raices = f'Raíces: {raices[0]:.2f}, {raices[1]:.2f}'

        print(f"📊 {info_raices}")

    else:
        print("📊 No hay raíces reales (discriminante < 0)")

    # Ajustar límites
    margen = 2
    y_min = min(-10, np.min(y) - margen)
    y_max = max(10, np.max(y) + margen)
    plt.xlim([-10, 10])
    plt.ylim([y_min, y_max])

    # Agregar leyenda
    plt.legend()
    plt.show()

    # Mostrar información adicional
    print(f"📐 Ecuación: f(x) = {a:.2f}x² + {b:.2f}x + {c:.2f}")
    if a != 0:
        x_v = -b / (2*a)
        y_v = f(x_v, a, b, c)
        print(f"📍 Vértice: ({x_v:.2f}, {y_v:.2f})")

    # Mostrar el discriminante
    discriminante = b**2 - 4*a*c
    print(f"📈 Discriminante: {discriminante:.2f}")
    if discriminante > 0:
        print("✅ Dos raíces reales diferentes")
    elif discriminante == 0:
        print("🔄 Una raíz real doble")
    else:
        print("❌ No hay raíces reales")
```

### Código para polinomios de tercer grado

```python
# Bloque 1: Importaciones y definición de funciones para polinomio de tercer grado
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Función del polinomio de tercer grado
def f3(x, a, b, c, d):
    """
    Calcula el valor del polinomio de tercer grado
    f(x) = ax³ + bx² + cx + d
    """
    return a * x**3 + b * x**2 + c * x + d

def derivada_polinomio3(a, b, c):
    """
    Calcula los coeficientes de la derivada del polinomio de tercer grado
    f'(x) = 3ax² + 2bx + c
    Retorna: (a_der, b_der, c_der)
    """
    return 3*a, 2*b, c

def encontrar_raices_polinomio3(a, b, c, d):
    """
    Encuentra las raíces reales de un polinomio de tercer grado usando numpy
    Retorna: lista de raíces reales
    """
    if a == 0:
        # Si a=0, es un polinomio de segundo grado
        if b == 0:
            return []
        # Resolver bx² + cx + d = 0
        discriminante = c**2 - 4*b*d
        if discriminante < 0:
            return []
        elif discriminante == 0:
            return [-c/(2*b)]
        else:
            return [(-c + np.sqrt(discriminante))/(2*b), (-c - np.sqrt(discriminante))/(2*b)]

    # Coeficientes del polinomio: ax³ + bx² + cx + d
    coeficientes = [a, b, c, d]

    # Encontrar todas las raíces (incluyendo complejas)
    raices = np.roots(coeficientes)

    # Filtrar solo raíces reales
    raices_reales = []
    for raiz in raices:
        if np.isreal(raiz):
            raiz_real = np.real(raiz)
            # Verificar que realmente es raíz (aproximadamente)
            if abs(f3(raiz_real, a, b, c, d)) < 1e-6:
                raices_reales.append(raiz_real)

    # Eliminar duplicados
    raices_reales = list(set(raices_reales))
    return sorted(raices_reales)

def encontrar_puntos_criticos3(a, b, c, d):
    """
    Encuentra los puntos críticos (máximos y mínimos) del polinomio de tercer grado
    Resolviendo f'(x) = 3ax² + 2bx + c = 0
    """
    if a == 0:
        return []

    a_der, b_der, c_der = derivada_polinomio3(a, b, c)

    # Resolver 3ax² + 2bx + c = 0
    discriminante = b_der**2 - 4*a_der*c_der

    if discriminante < 0:
        return []  # Sin puntos críticos reales
    elif discriminante == 0:
        x = -b_der / (2*a_der)
        y = f3(x, a, b, c, d)
        return [(x, y)]  # Punto de inflexión
    else:
        x1 = (-b_der + np.sqrt(discriminante)) / (2*a_der)
        x2 = (-b_der - np.sqrt(discriminante)) / (2*a_der)
        y1 = f3(x1, a, b, c, d)
        y2 = f3(x2, a, b, c, d)
        return [(x1, y1), (x2, y2)]

def graficar_polinomio3(a=1.0, b=0.0, c=0.0, d=0.0):
    """
    Función que grafica el polinomio de tercer grado con los coeficientes dados
    y marca las raíces en color morado
    """
    # Crear datos para graficar
    x = np.linspace(-10, 10, 500)
    y = f3(x, a, b, c, d)

    # Crear la gráfica
    plt.figure(figsize=(10, 6))

    # Graficar el polinomio
    plt.plot(x, y, linewidth=2, color='blue',
             label=f'f(x) = {a:.2f}x³ + {b:.2f}x² + {c:.2f}x + {d:.2f}')

    # Configurar el gráfico
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Polinomio de tercer grado: ax³ + bx² + cx + d')
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)

    # Encontrar y mostrar raíces (puntos morados)
    raices = encontrar_raices_polinomio3(a, b, c, d)

    # Graficar las raíces
    if raices:
        for i, raiz in enumerate(raices):
            # Verificar que la raíz esté dentro del rango de visualización
            if -10 <= raiz <= 10:
                y_raiz = f3(raiz, a, b, c, d)  # Esto debería ser ~0
                plt.plot(raiz, y_raiz, 'm^', markersize=10,
                        label='Raíces' if i == 0 else "")

                # Mostrar el valor de la raíz en el gráfico
                plt.annotate(f'x = {raiz:.2f}',
                           xy=(raiz, y_raiz),
                           xytext=(10, 10),
                           textcoords='offset points',
                           color='purple',
                           fontsize=10,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        # Información de raíces
        if len(raices) == 1:
            print(f"🟣 Raíz: {raices[0]:.2f}")
        elif len(raices) == 2:
            print(f"🟣 Raíces: {raices[0]:.2f}, {raices[1]:.2f}")
        else:
            print(f"🟣 Raíces: {raices[0]:.2f}, {raices[1]:.2f}, {raices[2]:.2f}")
    else:
        print("❌ No hay raíces reales")

    # Encontrar y mostrar puntos críticos (máximos/mínimos) en verde
    puntos_criticos = encontrar_puntos_criticos3(a, b, c, d)
    if puntos_criticos:
        for i, (x_c, y_c) in enumerate(puntos_criticos):
            if -10 <= x_c <= 10:
                # Determinar si es máximo o mínimo usando la segunda derivada
                # f''(x) = 6ax + 2b
                segunda_derivada = 6*a*x_c + 2*b
                if segunda_derivada > 0:
                    tipo = "Mínimo"
                    color = 'green'
                elif segunda_derivada < 0:
                    tipo = "Máximo"
                    color = 'red'
                else:
                    tipo = "Punto de inflexión"
                    color = 'orange'

                plt.plot(x_c, y_c, 'o', color=color, markersize=8,
                        label=f'{tipo}' if i == 0 else "")

                # Etiqueta del punto crítico
                plt.annotate(f'{tipo} ({x_c:.2f}, {y_c:.2f})',
                           xy=(x_c, y_c),
                           xytext=(10, -15) if y_c > 0 else (10, 15),
                           textcoords='offset points',
                           color=color,
                           fontsize=9,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        print(f"📍 Puntos críticos encontrados: {len(puntos_criticos)}")
    else:
        print("📍 No hay puntos críticos reales (derivada sin raíces reales)")

    # Ajustar límites
    margen = 2
    y_min = min(-10, np.min(y) - margen)
    y_max = max(10, np.max(y) + margen)
    plt.xlim([-10, 10])
    plt.ylim([y_min, y_max])

    # Agregar leyenda
    plt.legend()
    plt.show()

    # Mostrar información adicional
    print(f"\n📐 Ecuación: f(x) = {a:.2f}x³ + {b:.2f}x² + {c:.2f}x + {d:.2f}")

    # Mostrar el discriminante y características
    print(f"📈 Coeficiente principal: {a:.2f}")
    if a > 0:
        print("⬆️ La función tiende a +∞ cuando x → +∞")
    elif a < 0:
        print("⬇️ La función tiende a -∞ cuando x → +∞")
    else:
        print("⚠️ El coeficiente principal es 0, no es un polinomio cúbico")
```
