### Código para polinomios de segundo grado


```python
# Bloque 1: Importaciones y definición de funciones
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Función del polinomio
def f(x, a, b, c):
    """
    Calcula el valor del polinomio de segundo grado
    f(x) = ax² + bx + c
    """
    return a * x**2 + b * x + c

def calcular_raices(a, b, c):
    """
    Calcula las raíces del polinomio de segundo grado
    Retorna: (raiz1, raiz2) o (None, None) si no hay raíces reales
    """
    if a == 0:
        # Si a=0, es una ecuación lineal
        if b != 0:
            return [-c/b], None
        return None, None

    discriminante = b**2 - 4*a*c

    if discriminante < 0:
        # No hay raíces reales
        return None, None
    elif discriminante == 0:
        # Una raíz doble
        raiz = -b / (2*a)
        return raiz, raiz
    else:
        # Dos raíces reales diferentes
        raiz1 = (-b + np.sqrt(discriminante)) / (2*a)
        raiz2 = (-b - np.sqrt(discriminante)) / (2*a)
        return raiz1, raiz2

def graficar_polinomio(a=1.0, b=0.0, c=0.0):
    """
    Función que grafica el polinomio con los coeficientes dados
    y marca las raíces en color morado
    """
    # Crear datos para graficar
    x = np.linspace(-10, 10, 1000)
    y = f(x, a, b, c)

    # Crear la gráfica
    plt.figure(figsize=(10, 6))

    # Graficar el polinomio
    plt.plot(x, y, linewidth=2, color='blue', label=f'f(x) = {a:.2f}x² + {b:.2f}x + {c:.2f}')

    # Configurar el gráfico
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Polinomio de segundo grado: ax² + bx + c')
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)

    # Calcular y mostrar el vértice
    if a != 0:
        x_v = -b / (2*a)
        y_v = f(x_v, a, b, c)
        plt.plot(x_v, y_v, 'ro', markersize=8, label=f'Vértice: ({x_v:.2f}, {y_v:.2f})')

    # Calcular y mostrar las raíces (puntos morados)
    raiz1, raiz2 = calcular_raices(a, b, c)

    # Lista para almacenar las raíces
    raices = []
    if raiz1 is not None:
        raices.append(raiz1)
    if raiz2 is not None and raiz2 != raiz1:
        raices.append(raiz2)

    # Graficar las raíces
    if raices:
        for raiz in raices:
            # Verificar que la raíz esté dentro del rango de visualización
            if -10 <= raiz <= 10:
                y_raiz = f(raiz, a, b, c)  # Esto debería ser ~0
                plt.plot(raiz, y_raiz, 'm^', markersize=10,
                        label='Raíces' if raiz == raices[0] else "")

                # Mostrar el valor de la raíz en el gráfico
                plt.annotate(f'x = {raiz:.2f}',
                           xy=(raiz, y_raiz),
                           xytext=(10, 10),
                           textcoords='offset points',
                           color='purple',
                           fontsize=10,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        # Agregar información sobre las raíces
        if len(raices) == 1:
            info_raices = f'Raíz: {raices[0]:.2f} (doble)'
        else:
            info_raices = f'Raíces: {raices[0]:.2f}, {raices[1]:.2f}'

        print(f"📊 {info_raices}")

    else:
        print("📊 No hay raíces reales (discriminante < 0)")

    # Ajustar límites
    margen = 2
    y_min = min(-10, np.min(y) - margen)
    y_max = max(10, np.max(y) + margen)
    plt.xlim([-10, 10])
    plt.ylim([y_min, y_max])

    # Agregar leyenda
    plt.legend()
    plt.show()

    # Mostrar información adicional
    print(f"📐 Ecuación: f(x) = {a:.2f}x² + {b:.2f}x + {c:.2f}")
    if a != 0:
        x_v = -b / (2*a)
        y_v = f(x_v, a, b, c)
        print(f"📍 Vértice: ({x_v:.2f}, {y_v:.2f})")

    # Mostrar el discriminante
    discriminante = b**2 - 4*a*c
    print(f"📈 Discriminante: {discriminante:.2f}")
    if discriminante > 0:
        print("✅ Dos raíces reales diferentes")
    elif discriminante == 0:
        print("🔄 Una raíz real doble")
    else:
        print("❌ No hay raíces reales")
```

### Código para polinomios de tercer grado

```python
# Bloque 1: Importaciones y definición de funciones para polinomio de tercer grado
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Función del polinomio de tercer grado
def f3(x, a, b, c, d):
    """
    Calcula el valor del polinomio de tercer grado
    f(x) = ax³ + bx² + cx + d
    """
    return a * x**3 + b * x**2 + c * x + d

def derivada_polinomio3(a, b, c):
    """
    Calcula los coeficientes de la derivada del polinomio de tercer grado
    f'(x) = 3ax² + 2bx + c
    Retorna: (a_der, b_der, c_der)
    """
    return 3*a, 2*b, c

def encontrar_raices_polinomio3(a, b, c, d):
    """
    Encuentra las raíces reales de un polinomio de tercer grado usando numpy
    Retorna: lista de raíces reales
    """
    if a == 0:
        # Si a=0, es un polinomio de segundo grado
        if b == 0:
            return []
        # Resolver bx² + cx + d = 0
        discriminante = c**2 - 4*b*d
        if discriminante < 0:
            return []
        elif discriminante == 0:
            return [-c/(2*b)]
        else:
            return [(-c + np.sqrt(discriminante))/(2*b), (-c - np.sqrt(discriminante))/(2*b)]

    # Coeficientes del polinomio: ax³ + bx² + cx + d
    coeficientes = [a, b, c, d]

    # Encontrar todas las raíces (incluyendo complejas)
    raices = np.roots(coeficientes)

    # Filtrar solo raíces reales
    raices_reales = []
    for raiz in raices:
        if np.isreal(raiz):
            raiz_real = np.real(raiz)
            # Verificar que realmente es raíz (aproximadamente)
            if abs(f3(raiz_real, a, b, c, d)) < 1e-6:
                raices_reales.append(raiz_real)

    # Eliminar duplicados
    raices_reales = list(set(raices_reales))
    return sorted(raices_reales)

def encontrar_puntos_criticos3(a, b, c, d):
    """
    Encuentra los puntos críticos (máximos y mínimos) del polinomio de tercer grado
    Resolviendo f'(x) = 3ax² + 2bx + c = 0
    """
    if a == 0:
        return []

    a_der, b_der, c_der = derivada_polinomio3(a, b, c)

    # Resolver 3ax² + 2bx + c = 0
    discriminante = b_der**2 - 4*a_der*c_der

    if discriminante < 0:
        return []  # Sin puntos críticos reales
    elif discriminante == 0:
        x = -b_der / (2*a_der)
        y = f3(x, a, b, c, d)
        return [(x, y)]  # Punto de inflexión
    else:
        x1 = (-b_der + np.sqrt(discriminante)) / (2*a_der)
        x2 = (-b_der - np.sqrt(discriminante)) / (2*a_der)
        y1 = f3(x1, a, b, c, d)
        y2 = f3(x2, a, b, c, d)
        return [(x1, y1), (x2, y2)]

def graficar_polinomio3(a=1.0, b=0.0, c=0.0, d=0.0):
    """
    Función que grafica el polinomio de tercer grado con los coeficientes dados
    y marca las raíces en color morado
    """
    # Crear datos para graficar
    x = np.linspace(-10, 10, 500)
    y = f3(x, a, b, c, d)

    # Crear la gráfica
    plt.figure(figsize=(10, 6))

    # Graficar el polinomio
    plt.plot(x, y, linewidth=2, color='blue',
             label=f'f(x) = {a:.2f}x³ + {b:.2f}x² + {c:.2f}x + {d:.2f}')

    # Configurar el gráfico
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Polinomio de tercer grado: ax³ + bx² + cx + d')
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)

    # Encontrar y mostrar raíces (puntos morados)
    raices = encontrar_raices_polinomio3(a, b, c, d)

    # Graficar las raíces
    if raices:
        for i, raiz in enumerate(raices):
            # Verificar que la raíz esté dentro del rango de visualización
            if -10 <= raiz <= 10:
                y_raiz = f3(raiz, a, b, c, d)  # Esto debería ser ~0
                plt.plot(raiz, y_raiz, 'm^', markersize=10,
                        label='Raíces' if i == 0 else "")

                # Mostrar el valor de la raíz en el gráfico
                plt.annotate(f'x = {raiz:.2f}',
                           xy=(raiz, y_raiz),
                           xytext=(10, 10),
                           textcoords='offset points',
                           color='purple',
                           fontsize=10,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        # Información de raíces
        if len(raices) == 1:
            print(f"🟣 Raíz: {raices[0]:.2f}")
        elif len(raices) == 2:
            print(f"🟣 Raíces: {raices[0]:.2f}, {raices[1]:.2f}")
        else:
            print(f"🟣 Raíces: {raices[0]:.2f}, {raices[1]:.2f}, {raices[2]:.2f}")
    else:
        print("❌ No hay raíces reales")

    # Encontrar y mostrar puntos críticos (máximos/mínimos) en verde
    puntos_criticos = encontrar_puntos_criticos3(a, b, c, d)
    if puntos_criticos:
        for i, (x_c, y_c) in enumerate(puntos_criticos):
            if -10 <= x_c <= 10:
                # Determinar si es máximo o mínimo usando la segunda derivada
                # f''(x) = 6ax + 2b
                segunda_derivada = 6*a*x_c + 2*b
                if segunda_derivada > 0:
                    tipo = "Mínimo"
                    color = 'green'
                elif segunda_derivada < 0:
                    tipo = "Máximo"
                    color = 'red'
                else:
                    tipo = "Punto de inflexión"
                    color = 'orange'

                plt.plot(x_c, y_c, 'o', color=color, markersize=8,
                        label=f'{tipo}' if i == 0 else "")

                # Etiqueta del punto crítico
                plt.annotate(f'{tipo} ({x_c:.2f}, {y_c:.2f})',
                           xy=(x_c, y_c),
                           xytext=(10, -15) if y_c > 0 else (10, 15),
                           textcoords='offset points',
                           color=color,
                           fontsize=9,
                           bbox=dict(boxstyle='round,pad=0.3',
                                   facecolor='white',
                                   alpha=0.7))

        print(f"📍 Puntos críticos encontrados: {len(puntos_criticos)}")
    else:
        print("📍 No hay puntos críticos reales (derivada sin raíces reales)")

    # Ajustar límites
    margen = 2
    y_min = min(-10, np.min(y) - margen)
    y_max = max(10, np.max(y) + margen)
    plt.xlim([-10, 10])
    plt.ylim([y_min, y_max])

    # Agregar leyenda
    plt.legend()
    plt.show()

    # Mostrar información adicional
    print(f"\n📐 Ecuación: f(x) = {a:.2f}x³ + {b:.2f}x² + {c:.2f}x + {d:.2f}")

    # Mostrar el discriminante y características
    print(f"📈 Coeficiente principal: {a:.2f}")
    if a > 0:
        print("⬆️ La función tiende a +∞ cuando x → +∞")
    elif a < 0:
        print("⬇️ La función tiende a -∞ cuando x → +∞")
    else:
        print("⚠️ El coeficiente principal es 0, no es un polinomio cúbico")
```

### Código para teoremas

```python
# Bloque 1: Análisis completo de polinomios con teoremas fundamentales
import numpy as np
from sympy import symbols, Poly, sturm, real_roots
from ipywidgets import interact, FloatSlider
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

# ==================== FUNCIONES AUXILIARES ====================

def formatear_polinomio(coeficientes):
    """
    Formatea el polinomio como string para mostrar
    coeficientes: lista [a_n, a_{n-1}, ..., a_0]
    """
    if not coeficientes:
        return "0"

    # Limpiar ceros al inicio
    while coeficientes and abs(coeficientes[0]) < 1e-12:
        coeficientes = coeficientes[1:]

    if not coeficientes:
        return "0"

    grado = len(coeficientes) - 1
    partes = []

    for i, coef in enumerate(coeficientes):
        exp = grado - i
        if abs(coef) > 1e-12:
            if i == 0:
                signo = ""
            else:
                signo = " + " if coef > 0 else " - "

            coef_abs = abs(coef)

            if abs(coef_abs - 1) > 1e-12 or exp == 0:
                if exp > 1:
                    parte = f"{signo}{coef_abs:.2f}x^{exp}"
                elif exp == 1:
                    parte = f"{signo}{coef_abs:.2f}x"
                else:
                    parte = f"{signo}{coef_abs:.2f}"
            else:
                if exp > 1:
                    parte = f"{signo}x^{exp}"
                elif exp == 1:
                    parte = f"{signo}x"
                else:
                    parte = f"{signo}1"

            partes.append(parte)

    return "".join(partes)

def limpiar_coeficientes(coeficientes):
    """
    Limpia coeficientes muy pequeños y elimina ceros al inicio
    """
    # Redondear coeficientes muy pequeños
    coeficientes = [0 if abs(c) < 1e-12 else c for c in coeficientes]

    # Eliminar ceros al inicio
    while coeficientes and abs(coeficientes[0]) < 1e-12:
        coeficientes = coeficientes[1:]

    return coeficientes

# ==================== TEOREMA 1: FUNDAMENTAL DEL ÁLGEBRA ====================

def teorema_fundamental_algebra(coeficientes):
    """
    Teorema Fundamental del Álgebra:
    Todo polinomio de grado n tiene exactamente n raíces (reales o complejas)
    """
    # Limpiar coeficientes
    coeficientes = limpiar_coeficientes(coeficientes)

    if not coeficientes:
        return "El polinomio es nulo (todos los coeficientes son cero)"

    grado = len(coeficientes) - 1

    resultado = []
    resultado.append("📌 TEOREMA FUNDAMENTAL DEL ÁLGEBRA:")
    resultado.append(f"   → El polinomio tiene EXACTAMENTE {grado} raíces en total")
    resultado.append(f"   → Estas raíces pueden ser reales o complejas")
    resultado.append(f"   → Las raíces se cuentan con su multiplicidad")

    return "\n".join(resultado)

# ==================== TEOREMA 2: DESCARTES (REGLA DE LOS SIGNOS) ====================

def teorema_descartes(coeficientes):
    """
    Regla de los signos de Descartes:
    - Número de raíces reales positivas = cambios de signo (o menos en número par)
    - Número de raíces reales negativas = cambios de signo en P(-x)
    """
    # Limpiar coeficientes
    coeficientes = limpiar_coeficientes(coeficientes)

    if not coeficientes or len(coeficientes) < 2:
        return "No se puede aplicar la regla de Descartes (polinomio constante o nulo)"

    # Contar cambios de signo para raíces positivas
    cambios_pos = 0
    for i in range(len(coeficientes)-1):
        if coeficientes[i] * coeficientes[i+1] < 0:
            cambios_pos += 1

    # Contar cambios de signo para raíces negativas (P(-x))
    # Los signos de los coeficientes alternan: a_n, -a_{n-1}, a_{n-2}, ...
    coef_neg = []
    for i, coef in enumerate(coeficientes):
        if i % 2 == 0:  # Términos pares (x^0, x^2, x^4...)
            coef_neg.append(coef)
        else:  # Términos impares (x^1, x^3, x^5...)
            coef_neg.append(-coef)

    cambios_neg = 0
    for i in range(len(coef_neg)-1):
        if coef_neg[i] * coef_neg[i+1] < 0:
            cambios_neg += 1

    resultado = []
    resultado.append("📌 REGLA DE LOS SIGNOS DE DESCARTES:")

    if cambios_pos == 0:
        resultado.append(f"   → Raíces reales POSITIVAS: 0")
    else:
        resultado.append(f"   → Raíces reales POSITIVAS: {cambios_pos} (o menos en un número par)")
        if cambios_pos >= 2:
            posibles = [cambios_pos - 2*i for i in range(cambios_pos//2 + 1)]
            resultado.append(f"      Posibles cantidades: {posibles}")

    if cambios_neg == 0:
        resultado.append(f"   → Raíces reales NEGATIVAS: 0")
    else:
        resultado.append(f"   → Raíces reales NEGATIVAS: {cambios_neg} (o menos en un número par)")
        if cambios_neg >= 2:
            posibles = [cambios_neg - 2*i for i in range(cambios_neg//2 + 1)]
            resultado.append(f"      Posibles cantidades: {posibles}")

    return "\n".join(resultado)

# ==================== TEOREMA 3: RAÍCES RACIONALES ====================

def teorema_raices_racionales(coeficientes):
    """
    Teorema de las raíces racionales:
    Si p/q es una raíz racional, entonces p divide a a₀ y q divide a aₙ
    """
    # Limpiar coeficientes
    coeficientes = limpiar_coeficientes(coeficientes)

    if not coeficientes or len(coeficientes) < 2:
        return "No se puede aplicar el teorema de raíces racionales"

    # Coeficiente principal y término independiente
    a_n = coeficientes[0]
    a_0 = coeficientes[-1]

    resultado = []
    resultado.append("📌 TEOREMA DE LAS RAÍCES RACIONALES:")

    if abs(a_0) < 1e-12:
        resultado.append("   → x = 0 es una raíz (término independiente = 0)")
        return "\n".join(resultado)

    # Verificar que los coeficientes sean enteros
    no_enteros = False
    for c in coeficientes:
        if not float(c).is_integer():
            no_enteros = True
            break

    if no_enteros:
        resultado.append("   ⚠️ Los coeficientes no son todos enteros")
        resultado.append("   → El teorema se aplica mejor con coeficientes enteros")
        resultado.append("   → Se recomienda multiplicar por un factor para enterizar")
        return "\n".join(resultado)

    # Encontrar divisores
    a_n_int = int(abs(a_n))
    a_0_int = int(abs(a_0))

    divisores_p = []
    for i in range(1, a_0_int + 1):
        if a_0_int % i == 0:
            divisores_p.append(i)
            divisores_p.append(-i)

    divisores_q = []
    for i in range(1, a_n_int + 1):
        if a_n_int % i == 0:
            divisores_q.append(i)
            divisores_q.append(-i)

    # Generar posibles raíces
    posibles = []
    for p in divisores_p:
        for q in divisores_q:
            if q != 0:
                raiz = p / q
                # Verificar si realmente es raíz
                valor = np.polyval(coeficientes, raiz)
                if abs(valor) < 1e-10:
                    posibles.append(raiz)

    # Eliminar duplicados y ordenar
    posibles = sorted(list(set(posibles)))

    resultado.append(f"   → Coeficiente principal (aₙ): {a_n_int}")
    resultado.append(f"   → Término independiente (a₀): {a_0_int}")
    resultado.append(f"   → Divisores de a₀: {divisores_p}")
    resultado.append(f"   → Divisores de aₙ: {divisores_q}")

    if posibles:
        resultado.append(f"\n   ✅ Raíces racionales encontradas: {posibles}")
        resultado.append(f"   → Número de raíces racionales: {len(posibles)}")
    else:
        resultado.append(f"\n   ❌ No se encontraron raíces racionales exactas")

    return "\n".join(resultado)

# ==================== TEOREMA 4: STURM ====================

def teorema_sturm(coeficientes):
    """
    Teorema de Sturm:
    Cuenta el número exacto de raíces reales en un intervalo
    """
    # Limpiar coeficientes
    coeficientes = limpiar_coeficientes(coeficientes)

    if not coeficientes or len(coeficientes) < 2:
        return "No se puede aplicar el teorema de Sturm"

    resultado = []
    resultado.append("📌 TEOREMA DE STURM:")

    try:
        # Usar sympy para la secuencia de Sturm
        x = symbols('x')
        polinomio = Poly(coeficientes, x)

        # Obtener la secuencia de Sturm
        sturm_seq = sturm(polinomio)

        resultado.append(f"   → Número de polinomios en la secuencia de Sturm: {len(sturm_seq)}")

        # Encontrar raíces reales con sympy
        raices_reales = real_roots(polinomio)

        if raices_reales:
            resultado.append(f"   ✅ Raíces reales encontradas: {len(raices_reales)}")
            for i, raiz in enumerate(raices_reales, 1):
                raiz_val = float(raiz)
                resultado.append(f"      → x_{i} = {raiz_val:.6f}")
        else:
            resultado.append("   ❌ No se encontraron raíces reales")

        # Contar raíces en intervalos específicos
        intervalo_inf = -10
        intervalo_sup = 10
        raices_intervalo = [r for r in raices_reales if intervalo_inf <= float(r) <= intervalo_sup]

        resultado.append(f"\n   📊 Raíces en el intervalo [{intervalo_inf}, {intervalo_sup}]:")
        if raices_intervalo:
            resultado.append(f"      → Número de raíces: {len(raices_intervalo)}")
            for i, raiz in enumerate(raices_intervalo, 1):
                resultado.append(f"      → x_{i} = {float(raiz):.6f}")
        else:
            resultado.append("      → No hay raíces en este intervalo")

        # Demostración del teorema de Sturm (contar raíces usando la secuencia)
        resultado.append(f"\n   📐 Demostración del Teorema de Sturm:")
        resultado.append(f"      → La secuencia de Sturm permite contar el número exacto")
        resultado.append(f"      → de raíces reales en un intervalo [a, b]")
        resultado.append(f"      → V(a) - V(b) = número de raíces en (a, b)")
        resultado.append(f"      → Donde V(x) es el número de cambios de signo en la secuencia")

    except Exception as e:
        resultado.append(f"   ⚠️ Error al calcular la secuencia de Sturm: {str(e)}")
        resultado.append("   → Usando método numérico alternativo...")

        # Método alternativo con numpy
        raices = np.roots(coeficientes)
        raices_reales = [r for r in raices if np.isreal(r)]

        if raices_reales:
            resultado.append(f"   ✅ Raíces reales encontradas (método numérico): {len(raices_reales)}")
            for i, r in enumerate(raices_reales, 1):
                resultado.append(f"      → x_{i} = {np.real(r):.6f}")
        else:
            resultado.append("   ❌ No se encontraron raíces reales")

    return "\n".join(resultado)

# ==================== ANÁLISIS COMPLETO ====================

def analisis_completo_polinomio(a, b, c, d=0, e=0):
    """
    Función principal que aplica todos los teoremas
    """
    # Construir coeficientes del polinomio
    coeficientes = []

    # Agregar coeficientes en orden de mayor a menor grado
    if abs(e) > 1e-12:
        coeficientes = [a, b, c, d, e]
    elif abs(d) > 1e-12:
        coeficientes = [a, b, c, d]
    elif abs(c) > 1e-12:
        coeficientes = [a, b, c]
    elif abs(b) > 1e-12:
        coeficientes = [a, b]
    elif abs(a) > 1e-12:
        coeficientes = [a]
    else:
        return "Error: Todos los coeficientes son cero"

    # Limpiar coeficientes
    coeficientes = limpiar_coeficientes(coeficientes)

    if not coeficientes:
        return "Error: Polinomio nulo"

    grado = len(coeficientes) - 1
    polinomio_str = formatear_polinomio(coeficientes)

    # Construir el resultado
    resultado = []
    resultado.append("="*75)
    resultado.append(f"📊 ANÁLISIS COMPLETO DEL POLINOMIO")
    resultado.append(f"   f(x) = {polinomio_str}")
    resultado.append(f"   Grado: {grado}")
    resultado.append("="*75)

    # Aplicar todos los teoremas
    resultado.append("\n" + teorema_fundamental_algebra(coeficientes))
    resultado.append("\n" + teorema_descartes(coeficientes))
    resultado.append("\n" + teorema_raices_racionales(coeficientes))
    resultado.append("\n" + teorema_sturm(coeficientes))

    # Resumen final
    resultado.append("\n" + "="*75)
    resultado.append("📋 RESUMEN FINAL:")

    # Contar raíces reales numéricamente
    raices = np.roots(coeficientes)
    raices_reales = [r for r in raices if np.isreal(r)]
    raices_complejas = [r for r in raices if not np.isreal(r)]

    resultado.append(f"   • Grado del polinomio: {grado}")
    resultado.append(f"   • Coeficiente principal: {coeficientes[0]:.4f}")
    resultado.append(f"   • Término independiente: {coeficientes[-1]:.4f}")
    resultado.append(f"   • Número total de raíces: {len(raices)} (incluyendo complejas)")
    resultado.append(f"   • Raíces reales: {len(raices_reales)}")
    resultado.append(f"   • Raíces complejas: {len(raices_complejas)}")

    if raices_reales:
        resultado.append(f"\n   ✅ Raíces reales encontradas:")
        for i, r in enumerate(sorted(raices_reales), 1):
            resultado.append(f"      → x_{i} = {np.real(r):.6f}")

    if raices_complejas:
        resultado.append(f"\n   🔢 Raíces complejas:")
        for i, r in enumerate(raices_complejas, 1):
            resultado.append(f"      → z_{i} = {r:.6f}")

    resultado.append("="*75)

    # Mostrar teoremas aplicados
    resultado.append("\n📚 TEOREMAS APLICADOS:")
    resultado.append("   1. Teorema Fundamental del Álgebra")
    resultado.append("   2. Regla de los signos de Descartes")
    resultado.append("   3. Teorema de las raíces racionales")
    resultado.append("   4. Teorema de Sturm")
    resultado.append("="*75)

    return "\n".join(resultado)

# ==================== FUNCIÓN PARA MOSTRAR EN COLAB ====================

def mostrar_analisis_polinomio(a=1.0, b=0.0, c=0.0, d=0.0, e=0.0):
    """
    Función que muestra el análisis en formato HTML para Colab
    """
    # Limpiar output anterior
    from IPython.display import clear_output
    clear_output(wait=True)

    # Mostrar título
    display(HTML("<h2>📊 ANALIZADOR COMPLETO DE POLINOMIOS</h2>"))
    display(HTML("<h4>Aplicando: Teorema Fundamental, Descartes, Raíces Racionales y Sturm</h4>"))
    display(HTML("<hr>"))

    # Ejecutar análisis
    resultado = analisis_completo_polinomio(a, b, c, d, e)

    # Mostrar resultado en formato preformateado
    display(HTML(f"<pre style='font-size:12px; background-color:#f5f5f5; padding:15px; border-radius:5px;'>{resultado}</pre>"))

    # Mostrar información de los sliders
    display(HTML("<br><hr>"))
    display(HTML("<b>🔧 Ajusta los coeficientes usando los sliders:</b>"))

    # Mostrar qué coeficientes están activos según el grado
    coef_activos = []
    if abs(e) > 1e-12:
        coef_activos = ['a (x⁴)', 'b (x³)', 'c (x²)', 'd (x)', 'e (constante)']
    elif abs(d) > 1e-12:
        coef_activos = ['a (x³)', 'b (x²)', 'c (x)', 'd (constante)']
    elif abs(c) > 1e-12:
        coef_activos = ['a (x²)', 'b (x)', 'c (constante)']
    elif abs(b) > 1e-12:
        coef_activos = ['a (x)', 'b (constante)']
    else:
        coef_activos = ['a (constante)']

    display(HTML(f"<i>Coeficientes activos: {', '.join(coef_activos)}</i>"))
```

## Teoremas

### Teorema fundamental del álgebra

**Todo polinomio de grado n con coeficientes complejos tiene exactamente n raíces complejas, contando multiplicidades.**

Explicación:

- Un polinomio de grado 1 tiene 1 raíz
- Un polinomio de grado 2 tiene 2 raíces
- Un polinomio de grado 3 tiene 3 raíces

En general, un polinomio de grado n tiene n raíces

Importante:
- Las raíces pueden ser reales o complejas
- Las raíces se cuentan con su multiplicidad (una raíz doble cuenta como 2)
- Ejemplo: x² - 4x + 4 = 0 tiene raíz doble en x = 2 (cuenta como 2 raíces)

Ejemplos:
- x² + 1 = 0 → Raíces: i, -i (2 raíces complejas)
- x³ - 1 = 0 → Raíces: 1, -1/2 + i√3/2, -1/2 - i√3/2 (1 real, 2 complejas)
- x⁴ - 16 = 0 → Raíces: 2, -2, 2i, -2i (2 reales, 2 complejas)

### REGLA DE LOS SIGNOS DE DESCARTES

**El número de raíces reales positivas de un polinomio es igual al número de cambios de signo en la secuencia de sus coeficientes (ordenados de mayor a menor grado), o menos en un número par.**

Para las raíces reales negativas, se aplica la misma regla al polinomio P(-x).

Regla práctica:
- Raíces positivas: Cuenta los cambios de signo en los coeficientes de P(x)
- Raíces negativas: Evalúa P(-x) y cuenta los cambios de signo

Ejemplo con P(x) = x³ - 6x² + 11x - 6:

Coeficientes: +1, -6, +11, -6
Cambios de signo: +→- (1), -→+ (2), +→- (3)
→ 3 cambios de signo → 3 raíces positivas (o 1)

Para P(-x) = -x³ - 6x² - 11x - 6
Coeficientes: -1, -6, -11, -6
Cambios de signo: 0
→ 0 raíces negativas

Conclusión: Tiene 3 raíces positivas y 0 negativas (efectivamente: 1, 2, 3)

### TEOREMA DE LAS RAÍCES RACIONALES

**Si un polinomio con coeficientes enteros:**
P(x) = aₙxⁿ + aₙ₋₁xⁿ⁻¹ + ... + a₁x + a₀

**tiene una raíz racional p/q (en su mínima expresión), entonces:**
- **p divide al término independiente a₀**
- **q divide al coeficiente principal aₙ**

Cómo usarlo:
- Encuentra todos los divisores del término independiente (a₀)
- Encuentra todos los divisores del coeficiente principal (aₙ)
- Forma todas las fracciones p/q posibles
- Prueba cada una en el polinomio

Ejemplo con P(x) = 2x³ - 3x² - 8x + 12:
- a₀ = 12 → divisores: ±1, ±2, ±3, ±4, ±6, ±12
- aₙ = 2 → divisores: ±1, ±2

Posibles raíces racionales: ±1, ±2, ±3, ±4, ±6, ±12, ±1/2, ±3/2
- P(2) = 16 - 12 - 16 + 12 = 0 → x = 2 es raíz
- P(-3/2) = 2(-27/8) - 3(9/4) - 8(-3/2) + 12 = 0 → x = -3/2 es raíz

Limitaciones:
- Solo funciona con coeficientes enteros
- Solo encuentra raíces racionales (no irracionales ni complejas)
- Da posibles candidatos, no todos son raíces

### TEOREMA DE STURM

**El Teorema de Sturm permite determinar el número exacto de raíces reales distintas de un polinomio en un intervalo [a, b].**

Secuencia de Sturm:
Dado un polinomio P(x), se construye la secuencia:

P₀(x) = P(x)
P₁(x) = P'(x) (derivada)
P₂(x) = -resto(P₀, P₁)
P₃(x) = -resto(P₁, P₂)
...
Pₙ(x) = constante (último resto no nulo)


Procedimiento:
- Construir la secuencia de Sturm
- Evaluar cada polinomio en x = a y x = b
- Contar cambios de signo en cada evaluación
    V(a) - V(b) = número de raíces en (a, b)

Ejemplo con P(x) = x³ - 6x² + 11x - 6:

Secuencia de Sturm:
P₀ = x³ - 6x² + 11x - 6
P₁ = 3x² - 12x + 11
P₂ = (2/3)x - 2/3
P₃ = 1

Evaluando en a = -∞ y b = +∞:
V(-∞) = 3 cambios
V(+∞) = 0 cambios
→ 3 - 0 = 3 raíces reales

Ventajas del Teorema de Sturm:
- Da el número EXACTO de raíces reales
- Funciona para polinomios de cualquier grado
- No requiere encontrar las raíces
- Comparación con otros métodos:
- Más preciso que Descartes (que da solo una cota)
- No se limita a raíces racionales como el teorema de raíces racionales
- Da el número exacto, no solo posibles cantidades
