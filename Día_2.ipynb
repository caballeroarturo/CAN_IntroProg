```python
# (Lista)
# Es una colección ordenada y modificable (mutable) de elementos. Puede contener tipos de datos mezclados y duplicados. Se define con corchetes
# Una lista de calificaciones o índices

puntuaciones = [8.5, 9.0, 7.2, 9.5]

# Podemos modificar un elemento
puntuaciones[2] = 8.0

print(f"Tipo: {type(puntuaciones)} | Lista modificada: {puntuaciones}")
```

```python
# (Tupla)
# Es una colección ordenada pero inmutable (no se puede modificar una vez creada).
# Es ideal para datos que deben protegerse contra cambios accidentales (como coordenadas o configuraciones fijas).
# Se define con paréntesis ()

# Dimensiones de una matriz o coordenadas geográficas (Lat, Lon)
coordenadas_cimat = (21.018, -101.257)

print(f"Tipo: {type(coordenadas_cimat)} | Valor: {coordenadas_cimat}")
# Nota: Si intentas hacer coordenadas_cimat[0] = 22.0, Python lanzará un error.
```

Diccionario

Es una estructura que almacena datos en parejas de clave-valor (key-value). En lugar de acceder por una posición numérica, accedes mediante su clave. Se define con llaves {} y dos puntos :.

```python
# Estructura perfecta para representar un registro completo
alumno = {
    "id": 1024,
    "nombre": "Carlos",
    "materia": "Psicometría",
    "calificacion_final": 9.2
}

# Acceder a un dato específico mediante su clave
print(f"Tipo: {type(alumno)}")
print(f"El alumno {alumno['nombre']} obtuvo {alumno['calificacion_final']} en {alumno['materia']}.")
```

Conjunto
Es una colección no ordenada de elementos únicos (no permite duplicados). Es excelente para operaciones matemáticas de conjuntos como uniones, intersecciones o para eliminar duplicados de una lista de golpe. Se define con llaves {}.

```python
# Definimos elementos, incluyendo duplicados a propósito
ID_usuarios = {101, 102, 103, 101, 104, 102}

# Al imprimirlo, Python elimina automáticamente los duplicados
print(f"Tipo: {type(ID_usuarios)} | Elementos únicos: {ID_usuarios}")
```

## Datos predeterminados por el sistema y datos predeterminados por el usuario

**Datos predeterminados por el sistema (Built-in)**: Son los tipos de datos nativos que ya vienen integrados en el lenguaje de programación. El sistema sabe perfectamente cuánta memoria ocupan y qué operaciones se pueden hacer con ellos de forma automática.

**Datos predeterminados por el usuario (User-defined):** Son estructuras personalizadas que creas cuando los tipos nativos no son suficientes para modelar un objeto de la vida real. En Python, esto se logra utilizando Clases.

```python
# --- 1. Datos predeterminados por el sistema ---
edad = 35                         # Entero (int)
calificacion = 9.8                 # Decimal / Punto flotante (float)
nombre_puesto = "Análisis de Datos" # Texto / Cadena (str)
activo = True                      # Booleano (bool)

print("Tipos del sistema:")
print(f"edad: {type(edad)}, calificacion: {type(calificacion)}, puesto: {type(nombre_puesto)}\n")
```

```python
# --- 2. Datos predeterminados por el usuario ---
# Definimos una estructura propia usando una clase
class Candidato:
    def __init__(self, nombre, puesto, experiencia_anios):
        self.nombre = nombre
        self.puesto = puesto
        self.experiencia_anios = experiencia_anios

# Creamos una instancia de nuestro propio tipo de dato
postulante = Candidato(nombre="Alice", puesto="Subgerente", experiencia_anios=15)

print("Tipo definido por el usuario:")
print(f"Tipo del objeto 'postulante': {type(postulante)}")
print(f"Nombre: {postulante.nombre}, Puesto: {postulante.puesto}, Experiencia: {postulante.experiencia_anios} años")
```

# Constantes
Una constante es un valor que se asigna al inicio y no debe cambiar a lo largo de la ejecución del programa (por ejemplo, el valor de $\pi$ o el porcentaje del IVA).Nota importante sobre Python: Python no tiene una palabra clave estricta para bloquear constantes (como const en JavaScript o final en Java). En su lugar, la comunidad de programadores utiliza una convención de nomenclatura: escribimos el nombre de la variable completamente en MAYÚSCULAS para avisar a otros desarrolladores que ese valor es intocable.

```python
# Se podría usar esta convención: Nombres en MAYÚSCULAS para constantes
PI = 3.1415926535
LIMITE_INTENTOS_LOGIN = 5
URL_API_DATOS = "https://api.analitica.internal/v1"

# Uso teórico
radio = 5
area = PI * (radio ** 2)

print(f"El área calculada usando la constante PI es: {area}")

# Nota: Python te permitirá hacer "PI = 4.0" sin lanzar error,
# pero hacer eso rompería las buenas prácticas de programación.
```

## Variables y expresiones

**Variable:** Es un espacio en la memoria del computador con un nombre asignado que sirve para almacenar un valor. A diferencia de las constantes, su contenido sí puede cambiar o variar durante la ejecución. Python es de tipado dinámico, lo que significa que no necesitas declarar qué tipo de dato guardará antes de usarla.

**Expresión:** Es una combinación de variables, valores, operadores y llamadas a funciones que el lenguaje procesa para evaluar y producir un nuevo resultado. Por ejemplo, x + 10 es una expresión.

```python
# --- Variables ---
# El valor cambia a lo largo del tiempo
registros_procesados = 150
registros_procesados = registros_procesados + 25  # El valor se actualiza

print(f"Valor actualizado de la variable: {registros_procesados}")
```

```python

# --- Expresiones ---
# Las expresiones se evalúan para generar un resultado final
pago_por_hora = 45.5
horas_trabajadas = 40

# Esta operación matemática es una expresión:
sueldo_semanal = horas_trabajadas * pago_por_hora

# Esta comparación lógica también es una expresión (evalúa a True o False):
aplica_bono = horas_trabajadas > 35

print(f"Resultado de expresión (Sueldo): {sueldo_semanal}")
print(f"Resultado de expresión lógica (Bono): {aplica_bono}")
```

# Entrada de datos

Es el mecanismo mediante el cual el programa recibe información desde el exterior, usualmente a través del teclado del usuario, de un archivo de texto o de una base de datos.

En Python, la función estándar para interactuar con el teclado es input(). Un detalle crítico que siempre debes recordar: ***input() siempre captura la información como texto (string)***. Si necesitas que el dato sea un número, debes transformarlo explícitamente (hacer un cast).

```python
# Captura de datos de texto
usuario = input("Por favor, ingresa tu nombre: ")

# Captura de datos numéricos (requiere conversión)
# int() convierte el texto introducido (ej. "5") en un número entero real (5)
anios_experiencia = int(input("Ingresa tus años de experiencia: "))

# Procesando las entradas
proximo_anio_experiencia = anios_experiencia + 1

print("\n--- Datos Procesados ---")
print(f"Hola {usuario}, el próximo año cumplirás {proximo_anio_experiencia} años de experiencia.")
```

## Salida de información

La salida es la forma en que el programa proyecta los resultados de su procesamiento hacia el mundo exterior (pantalla del usuario, un reporte en Excel, una gráfica o un archivo de texto).

En Python, la herramienta principal es la función print(). Para presentar datos de manera limpia, profesional y legible, se utilizan las f-strings (cadenas formateadas). Colocando una f antes de las comillas, puedes meter variables directamente dentro del texto usando llaves {}.

```python
# Datos de ejemplo obtenidos de un supuesto análisis
metrica = "R-cuadrado (R²)"
valor_metrica = 0.94582

# Salida estándar o básica
print("Salida Básica:")
print("Métrica:", metrica, "Valor:", valor_metrica)

print("\n-------------------------------------------")

# Salida avanzada con f-strings (Muy útil para reportes de datos)
# Podemos limitar los decimales sobre la marcha. Por ejemplo, a 3 decimales usando ':.3f'
print("Salida Formateada:")
print(f"El indicador {metrica} del modelo de regresión es exactamente {valor_metrica:.3f}.")
```
